```python
import numpy as np
num_Data=['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']
for i in num_Data:
    data[i]=np.log(data[i])
```

```python
for i in catagirical_data:
    temp=data.groupby(i)['SalePrice'].count()/len(data)
    temp_data=temp[temp>0.1].index
    data[i]=np.where(data[i].isin(temp_data),data[i],'Rare_Var')
```

```python
data.head()
```

```python
for i in catagirical_data:
    lavel=data.groupby([i])['SalePrice'].mean().sort_values().index
    lavel={k:j for j,k in enumerate(lavel,0)}
    data[i]=data[i].map(lavel)
```

```python
data=data.drop(['Id'],axis=1)
```

```python
final_Data=data.iloc[:,:-1]
y=data.iloc[:,-1:]
```

```python
final_Data.head()
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
select_model=SelectFromModel(Lasso(alpha=0.005,random_state=0))
select_model.fit(final_Data,y)
```

```python
select_model.get_support()
```

```python
feature=final_Data.columns[(select_model.get_support())]
print(len(feature))
```

```python
data[feature].head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
import pickle
```

```python
x_train,x_test,y_train,y_test=train_test_split(data[feature],y,test_size=0.33,random_state=42)
```

```python
def scaled(x_train,x_test):
    scaler=StandardScaler()    
    x_train_scaled=scaler.fit_transform(x_train)
    x_test_scaled=scaler.transform(x_test)
    
    file=open('Scalar.pkl','wb')
    pickle.dump(scaler,file)
    file.close()
    
    return x_train_scaled,x_test_scaled
```

```python
x_train_Scaled,x_test_scaled=scaled(x_train,x_test)
```

```python
x_train_Scaled
```

```python
x_test_scaled
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression
```

```python
regression.fit(x_train_Scaled,y_train)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
y_pred=regression.predict(x_test_scaled)
```

```python
y_pred
```

```python
## PErformance metrics cost function
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
file=open('House_Price.pkl','wb')
pickle.dump(regression,file)
file.close()
    
```
